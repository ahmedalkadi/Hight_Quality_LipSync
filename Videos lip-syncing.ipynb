# The Inference !

### In this section i will be using the wav2lip_gen and the wav2lip models for all the voices

```python
# list all the links of the videos 

base_url = '/kaggle/working/voices'
all_voices_links = []
for item in os.listdir(base_url):
    item_path = os.path.join(base_url, item)
    all_voices_links.append(str(item_path))
    
```

```python
# change the divectory to the wav2lip folder   
%cd Wav2Lip
```

## Using the Wav2Lip_gen.pth model

```python
# create a new folder for the wav2lip_gen generated videos
os.makedirs('/kaggle/working/wav2lip_gen_videos', exist_ok=True)

# define the path to the Wav2Lip GAN model
checkpoint_path = "/kaggle/working/models_folder/wav2lip_gan.pth"
face_path = "/kaggle/working/videos/face.mp4"
output_dir = "/kaggle/working/wav2lip_gen_videos/"

# loop through each audio file in all_voices_links
for i in all_voices_links:
    audio_path = i
    output_path = os.path.join(output_dir, f"{os.path.basename(i).split('.')[0]}_gen_model.mp4")
    
    command = (
        f'python "/kaggle/working/Wav2Lip/inference.py" '
        f'--checkpoint_path "{checkpoint_path}" '
        f'--face "{face_path}" '
        f'--audio "{audio_path}" '
        f'--outfile "{output_path}" '
        f'--pads 0 0 0 0'
    )
# -- notrim default value in False so in case of any silence in the begining or the end of the audio track it will be trimmed 

    subprocess.call(command, shell=True)
```

## Using the Wav2Lip.pth model

```python
# create a new folder for the wav2lip_gen generated videos
os.makedirs('/kaggle/working/wav2lip_videos', exist_ok=True)

# define the path to the Wav2Lip model
checkpoint_path = "/kaggle/working/models_folder/wav2lip.pth"
face_path = "/kaggle/working/videos/face.mp4"
output_dir = "/kaggle/working/wav2lip_videos/"

# loop through each audio file in all_voices_links
for i in all_voices_links:
    audio_path = i
    output_path = os.path.join(output_dir, f"{os.path.basename(i).split('.')[0]}_not_gen_model.mp4")
    
    command = (
        f'python "/kaggle/working/Wav2Lip/inference.py" '
        f'--checkpoint_path "{checkpoint_path}" '
        f'--face "{face_path}" '
        f'--audio "{audio_path}" '
        f'--outfile "{output_path}" '
        f'--pads 0 0 0 0'
    )
# -- notrim default value in False so in case of any silence in the begining or the end of the audio track it will be trimmed 
    
    subprocess.call(command, shell=True)
```
